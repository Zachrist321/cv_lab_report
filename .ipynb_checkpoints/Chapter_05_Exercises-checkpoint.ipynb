```python
%pip install -q opencv-python matplotlib numpy scikit-image scipy

import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from skimage import data, exposure, morphology, filters, measure, segmentation, util
from skimage.filters import sobel, prewitt
from skimage.segmentation import clear_border
from scipy import ndimage

STUDENT_NAME = "Zerihun Ayele Chala"
STUDENT_ID = "GSR/0979/18"

def find_images_dir() -> Path:
    for root in [Path.cwd(), Path("/content"), Path("/content/Computer vison (DIP) Lab ")]:
        d = root / "Images"
        if d.is_dir() and any(d.glob("*.jpg")):
            return d.resolve()
    raise FileNotFoundError("Upload the Images/ folder next to this notebook.")

IMAGES_DIR = find_images_dir()

def load_color(name="charlie-dt-XmlEVg-1nRk-unsplash.jpg"):
    path = IMAGES_DIR / name
    img = cv2.imread(str(path))
    if img is None:
        raise FileNotFoundError(path)
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def show(img, title="", cmap=None, figsize=(6, 4)):
    plt.figure(figsize=figsize)
    if cmap:
        plt.imshow(img, cmap=cmap)
    elif img.ndim == 2:
        plt.imshow(img, cmap="gray")
    else:
        plt.imshow(img)
    plt.title(title)
    plt.axis("off")
    plt.show()

print(f"Student: {STUDENT_NAME} | ID: {STUDENT_ID}")
print(f"Images: {IMAGES_DIR}")
```

```python
img = util.img_as_float(data.camera())
```

### Exercise 1 — Box filters 3×3, 5×5, 7×7

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
for a, k in zip(ax, [3, 5, 7]):
    ker = np.ones((k, k)) / (k * k)
    a.imshow(ndimage.convolve(img, ker), cmap="gray")
    a.set_title(f"{k}x{k} average"); a.axis("off")
plt.tight_layout(); plt.show()
```

### Exercise 2 — Unsharp masking

```python
blurred = filters.gaussian(img, sigma=1.5)
unsharp = np.clip(img + 1.5 * (img - blurred), 0, 1)
show(unsharp, "Unsharp Masking")
```

### Exercise 3 — PSNR on noisy median filtering

```python
from skimage.metrics import peak_signal_noise_ratio as psnr

rng = np.random.default_rng(1)
noisy = np.clip(img + rng.normal(0, 0.05, img.shape), 0, 1)
denoised = ndimage.median_filter(noisy, size=3)
print(f"PSNR noisy:  {psnr(img, noisy, data_range=1):.2f} dB")
print(f"PSNR median: {psnr(img, denoised, data_range=1):.2f} dB")
show(denoised, "Median denoised")
```

### Exercise 4 — Template correlation patch

```python
patch = img[100:140, 100:140]
corr = ndimage.correlate(img, patch - patch.mean(), mode='constant')
y, x = np.unravel_index(np.argmax(corr), corr.shape)
show(corr, f"Correlation map (peak at {x},{y})")
```
