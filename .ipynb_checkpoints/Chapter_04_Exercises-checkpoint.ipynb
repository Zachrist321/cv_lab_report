```python
%pip install -q opencv-python matplotlib numpy scikit-image scipy

import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from skimage import data, exposure, morphology, filters, measure, segmentation, util
from skimage.filters import sobel, prewitt
from skimage.segmentation import clear_border
from scipy import ndimage

STUDENT_NAME = "Zerihun Ayele Chala"
STUDENT_ID = "GSR/0979/18"

def find_images_dir() -> Path:
    for root in [Path.cwd(), Path("/content"), Path("/content/Computer vison (DIP) Lab ")]:
        d = root / "Images"
        if d.is_dir() and any(d.glob("*.jpg")):
            return d.resolve()
    raise FileNotFoundError("Upload the Images/ folder next to this notebook.")

IMAGES_DIR = find_images_dir()

def load_color(name="charlie-dt-XmlEVg-1nRk-unsplash.jpg"):
    path = IMAGES_DIR / name
    img = cv2.imread(str(path))
    if img is None:
        raise FileNotFoundError(path)
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def show(img, title="", cmap=None, figsize=(6, 4)):
    plt.figure(figsize=figsize)
    if cmap:
        plt.imshow(img, cmap=cmap)
    elif img.ndim == 2:
        plt.imshow(img, cmap="gray")
    else:
        plt.imshow(img)
    plt.title(title)
    plt.axis("off")
    plt.show()

print(f"Student: {STUDENT_NAME} | ID: {STUDENT_ID}")
print(f"Images: {IMAGES_DIR}")
```

```python
gray = data.coins()
binary = gray > filters.threshold_otsu(gray)
se = morphology.square(3)
```

### Exercise 1 — Different structuring elements

```python
elements = {"square3": morphology.square(3), "disk3": morphology.disk(3), "diamond3": morphology.diamond(3)}
fig, ax = plt.subplots(1, len(elements), figsize=(12, 4))
for a, (name, se) in zip(ax, elements.items()):
    a.imshow(morphology.binary_opening(binary, se), cmap="gray")
    a.set_title(name); a.axis("off")
plt.tight_layout(); plt.show()
```

### Exercise 2 — Opening/Closing on noisy binary image

```python
rng = np.random.default_rng(0)
noisy = binary ^ (rng.random(binary.shape) < 0.05)
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].imshow(noisy, cmap="gray"); ax[0].set_title("Noisy")
ax[1].imshow(morphology.binary_opening(noisy, se), cmap="gray"); ax[1].set_title("Opening")
ax[2].imshow(morphology.binary_closing(noisy, se), cmap="gray"); ax[2].set_title("Closing")
for a in ax: a.axis("off")
plt.tight_layout(); plt.show()
```

### Exercise 3 — Extract bright text-like regions (top-hat)

```python
room = load_color("clay-banks-nl_jlechVJA-unsplash.jpg")
g = cv2.cvtColor(room, cv2.COLOR_RGB2GRAY)
bw = g > filters.threshold_otsu(g)
text_like = morphology.white_tophat(bw, morphology.rectangle(15, 40))
show(text_like, "Top-hat feature extraction")
```
