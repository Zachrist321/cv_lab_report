```python
%pip install -q opencv-python matplotlib numpy scikit-image scipy

import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from skimage import data, exposure, morphology, filters, measure, segmentation, util
from skimage.filters import sobel, prewitt
from skimage.segmentation import clear_border
from scipy import ndimage

STUDENT_NAME = "Zerihun Ayele Chala"
STUDENT_ID = "GSR/0979/18"

def find_images_dir() -> Path:
    for root in [Path.cwd(), Path("/content"), Path("/content/Computer vison (DIP) Lab ")]:
        d = root / "Images"
        if d.is_dir() and any(d.glob("*.jpg")):
            return d.resolve()
    raise FileNotFoundError("Upload the Images/ folder next to this notebook.")

IMAGES_DIR = find_images_dir()

def load_color(name="charlie-dt-XmlEVg-1nRk-unsplash.jpg"):
    path = IMAGES_DIR / name
    img = cv2.imread(str(path))
    if img is None:
        raise FileNotFoundError(path)
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

def show(img, title="", cmap=None, figsize=(6, 4)):
    plt.figure(figsize=figsize)
    if cmap:
        plt.imshow(img, cmap=cmap)
    elif img.ndim == 2:
        plt.imshow(img, cmap="gray")
    else:
        plt.imshow(img)
    plt.title(title)
    plt.axis("off")
    plt.show()

print(f"Student: {STUDENT_NAME} | ID: {STUDENT_ID}")
print(f"Images: {IMAGES_DIR}")
```

### Exercise 1 — Otsu on multiple images

```python
tests = {
    "camera": data.camera(),
    "coins": data.coins(),
    "cycling": cv2.cvtColor(cv2.imread(str(IMAGES_DIR/"charlie-dt-XmlEVg-1nRk-unsplash.jpg")), cv2.COLOR_BGR2GRAY),
}
fig, ax = plt.subplots(len(tests), 2, figsize=(8, 10))
for i, (name, im) in enumerate(tests.items()):
    t = filters.threshold_otsu(im)
    ax[i,0].imshow(im, cmap="gray"); ax[i,0].set_title(name); ax[i,0].axis("off")
    ax[i,1].imshow(im > t, cmap="gray"); ax[i,1].set_title(f"Otsu T={t:.0f}"); ax[i,1].axis("off")
plt.tight_layout(); plt.show()
```

### Exercise 2 — Compare Sobel, Prewitt, Canny

```python
im = data.camera()
fig, ax = plt.subplots(1, 4, figsize=(14, 3))
ax[0].imshow(im, cmap="gray"); ax[0].set_title("Original"); ax[0].axis("off")
ax[1].imshow(sobel(im), cmap="gray"); ax[1].set_title("Sobel"); ax[1].axis("off")
ax[2].imshow(prewitt(im), cmap="gray"); ax[2].set_title("Prewitt"); ax[2].axis("off")
ax[3].imshow(cv2.Canny(im.astype(np.uint8), 50, 150), cmap="gray"); ax[3].set_title("Canny"); ax[3].axis("off")
plt.tight_layout(); plt.show()
```

### Exercise 3 — Region growing from seed

```python
from skimage.segmentation import flood

coins = data.coins()
seed = (coins.shape[0]//2, coins.shape[1]//2)
grown = flood(coins, seed, tolerance=15)
show(grown, f"Region Growing from seed {seed}")
```
